
About Dataset
Welcome to the "Worldwide Coffee Habits" dataset, a comprehensive resource designed to shed light on global coffee consumption trends. This dataset offers a detailed look into coffee consumption patterns across various countries, providing valuable insights into how coffee is enjoyed around the world.

The dataset includes the following features:

Country: The name of the country where the data was collected.
Year: The year of the record, spanning from 2000 to 2023.
Coffee Consumption (kg per capita per year): The amount of coffee consumed per person annually, offering a glimpse into the coffee habits of different nations.
Average Coffee Price (USD per kg): The average price of coffee per kilogram in US dollars, reflecting economic aspects of coffee consumption.
Type of Coffee Consumed: Insights into the most popular types of coffee enjoyed in each country, such as Espresso, Latte, Cappuccino, Americano, and Mocha.
Population (millions): The estimated population of each country, providing context for coffee consumption figures.

This dataset is a valuable asset for those interested in exploring how coffee consumption varies across different regions and over time. Whether you're a data scientist, market analyst, or coffee enthusiast, this dataset offers rich opportunities for analysis, from studying regional preferences and pricing trends to understanding the broader economic impacts of coffee consumption.

Dive into the data and uncover trends, patterns, and insights that highlight the diverse ways people around the globe enjoy their coffee.

```python
def simple_lineplot(title, df, x, y):
    plt.figure(figsize=(10,6))
    plt.title(title)
    sns.lineplot(data=df, x=x, y=y, marker='o')
    plt.xticks(df[x].unique(), rotation=45)
    plt.grid(True)
    plt.show()
```

```python
year_mean = lambda data, col, name: df.groupby('Year')[col].mean().reset_index(name=name)

consumption_years = year_mean(df, 'Coffee Consumption (kg per capita per year)', 'Consumption')
simple_lineplot('Average Coffee Consumption through the Years', consumption_years, 'Year', 'Consumption')
```

```python
price_years = year_mean(df, 'Average Coffee Price (USD per kg)', 'Price')
simple_lineplot('Average Coffee Price through the Years', price_years, 'Year', 'Price')
```

```python
for coffee_type in df['Type of Coffee Consumed'].unique():
    data = df[df['Type of Coffee Consumed'] == coffee_type]
    
    data_consumption = year_mean(data, 'Coffee Consumption (kg per capita per year)', 'Consumption')
    data_price = year_mean(data, 'Average Coffee Price (USD per kg)', 'Price')
    
    simple_lineplot(f'{coffee_type} Average Coffee Consumption through the Years', data_consumption, 'Year', 'Consumption')
    simple_lineplot(f'{coffee_type} Average Coffee Price through the Years', data_price, 'Year', 'Price')
```
